```python
dataset = w.get_waldos("256")
```

```python

@tf.function
def prepare_dataset(image_name: str , bounding_box: np.ndarray):
    image = tf.io.read_file(image_name)
    image = tf.image.decode_jpeg(image, channels=3)
    image = tf.image.resize(image, [227,227])
    
    return image, { "bbox": bounding_box }
# @tf.function
# def prepare_dataset(image_name: str , bounding_box: np.ndarray):


@tf.function
def adjust_brightness(image, labels):
    return tf.image.random_brightness(image, 19), labels





@tf.function
def flip(image, labels):
    bbox = labels["bbox"]
    img = tf.image.flip_left_right(image)
    lbl = { "bbox": [ 1 - bbox[0], 1 - bbox[1], 1 - bbox[2], 1 - bbox[3] ]}
    return img, lbl
    
    

    




```

```python
img, bbox  = [str(img) for img, bbox in dataset], [bbox.flatten().tolist() for img, bbox in dataset]
bbox = np.divide(np.array(bbox, dtype=np.float64), 256, dtype=np.float64)
print(bbox)

train_dataset = tf.data.Dataset.from_tensor_slices((img[0:15], bbox[0:15]))
train_dataset = train_dataset.map(prepare_dataset)
train_dataset.concatenate(train_dataset.map(adjust_brightness))
train_dataset.concatenate(train_dataset.map(flip))
train_dataset = train_dataset.shuffle(buffer_size=1024).batch(30).prefetch(AUTO)

#  


validation_dataset = tf.data.Dataset.from_tensor_slices((img[15:20], bbox[15:20]))
validation_dataset = validation_dataset.map(prepare_dataset)
validation_dataset = validation_dataset.shuffle(buffer_size=1024).batch(30).prefetch(AUTO)



test_dataset = tf.data.Dataset.from_tensor_slices((img[20:25], bbox[20:25]))
test_dataset = test_dataset.map(prepare_dataset)
test_dataset = test_dataset.batch(30).prefetch(AUTO)





```

```python
def get_yolo_model() -> Model: 
    model = Conv2D(64, (7,7), strides=(2, 2), activation="relu")
    model = MaxPooling2D((2,2), strides=(2, 2))(model)
    model = Conv2D(192, (3,3), activation="relu")(model)
    model = MaxPooling2D((2,2), strides=(2, 2))(model)
    model = Conv2D(128, (1,1), activation="relu")(model)
    model = Conv2D(256, (3,3), activation="relu")(model)
    model = Conv2D(256, (1,1), activation="relu")(model)
    model = Conv2D(512, (3,3), activation="relu")(model)
    model = MaxPooling2D((2,2), strides=(2, 2))(model)
    model = Conv2D(256, (1,1), activation="relu")(model)
    model = Conv2D(512, (3,3), activation="relu")(model)
    model = Conv2D(256, (1,1), activation="relu")(model)
    model = Conv2D(512, (3,3), activation="relu")(model)
    model = Conv2D(1024, (3,3), activation="relu", strides=(2, 2))(model)
    model = Conv2D(1024, (3,3), activation="relu")(model)
    model = Dense(1024, activation="relu")(model)
    model = Dense(30, activation="relu")(model)
    return Model(outputs=[model])
    
    
    
```

```python

def get_faster_r_cnn_model() -> Model:
    inputs = Input(shape=(227, 227, 3))
    model = Conv2D(64, (3,3), activation="relu")(inputs)
    model = Conv2D(64, (3,3), activation="relu")(model)
    model = MaxPooling2D((2,2), strides=(2, 2))(model)
    model = Conv2D(128, (3,3), activation="relu")(model)
    model = Conv2D(128, (3,3), activation="relu")(model)
    model = MaxPooling2D((2,2), strides=(2, 2))(model)
    model = Conv2D(256, (3,3), activation="relu")(model)
    model = Conv2D(256, (3,3), activation="relu")(model)
    model = Conv2D(256, (3,3), activation="relu")(model)
    model = MaxPooling2D((2,2), strides=(2, 2))(model)
    model = Conv2D(512, (3,3), activation="relu")(model)
    model = Conv2D(512, (3,3), activation="relu")(model)
    model = Conv2D(512, (3,3), activation="relu")(model)
    model = MaxPooling2D((2,2), strides=(2, 2))(model)
    model = Conv2D(512, (3,3), activation="relu")(model)
    model = Conv2D(512, (3,3), activation="relu")(model)
    model = Conv2D(512, (3,3), activation="relu")(model)
    model = MaxPooling2D((2,2), strides=(2, 2))(model)
    model = Dense(4096, activation="relu")(model)
    model = Dense(4096, activation="relu")(model)
    model = Dense(4096, activation="relu")(model)
    return Model(inputs=[inputs], outputs=[model])
```

```python
def get_bb_model() -> Model:
    model = get_faster_r_cnn_model()
    # model = VGG16(weights="imagenet", include_top=False,input_tensor=Input(shape=(224, 224, 3)))
    # model.trainable = False
    # flatten = Flatten()(model.output)
    bboxHead = Dense(128, activation="relu")(model.output)
    bboxHead = Dense(64, activation="relu")(bboxHead)
    bboxHead = Dense(32, activation="relu")(bboxHead)
    bboxHead = Dense(4, name="bbox")(bboxHead)
    return Model(inputs=model.input, outputs=[bboxHead])
```

```python
losses = {
"bbox": "mean_squared_error",
}
# define a dictionary that specifies the weights per loss (both the
# class label and bounding box outputs will receive equal weight)
lossWeights = {
"bbox": 1.0
}

model = get_yolo_model()
model.compile(optimizer=Adam(learning_rate=0.000001),
    loss=losses,
    metrics=['accuracy'],
    loss_weights=lossWeights)

# lbl = to_categorical(training_labels)
model.fit(train_dataset, epochs=20, validation_batch_size=validation_dataset)

print()
print(bbox[25:28])
```
